```python
def testEV1(lam):
    N = len(lam)
    lhs = DET(sum(u[i]*chi2d(lam[i]) for i in range(N)))*evaluate_on_tensor(omega_on_tensor(iGMP(lam)),[epsilon([])*q3^i for i in range(N)])
    rhs = evaluate_on_tensor(iGMP(lam),[u[N-i-1]*epsilon([]) for i in range(N)])
    return lhs == rhs
```

```python
def testEV2(lam):
    N = len(lam)
    lhs = DET(sum(u[i]*chi2d(lam[i]) for i in range(N)))*evaluate_on_tensor(omega_on_tensor(GMP(lam)),[epsilon([])*(v[i]-(1-q3)*sum(q3^(i-j-1)*v[j] for j in range(i))) for i in range(N)])
    rhs = evaluate_on_tensor(GMP(lam),[u[i]*v[i]*epsilon([]) for i in range(N)])
    return lhs == rhs
```

```python
def testEV3(lam):
    N = len(lam)
    lhs = PE(-z*sum(u[i]*chi2d(lam[i]) for i in range(N))) * evaluate_on_tensor(iGMP(lam),[-epsilon([])*q3^i for i in range(N)])
    rhs = evaluate_on_tensor(iGMP(lam),[-epsilon([])*(q3^i-z*u[N-i-1]) for i in range(N)])
    return lhs == rhs
```

```python
def testEV4(lam):
    N = len(lam)
    lhs = PE(-z*sum(u[i]*chi2d(lam[i]) for i in range(N))) * evaluate_on_tensor(GMP(lam),[-epsilon([])*q3^i for i in range(N)])
    rhs = evaluate_on_tensor(GMP(lam),[-epsilon([])*(q3^i-z*u[i]) for i in range(N)])
    return lhs == rhs
```

```python
all(testEV4(lam) for N in range(1,5) for d in range(5) for lam in mPartitions(N,d))
```

```python
def testBH(N,n):
    if n<=0 or N<=0:
        return True
    X = [tensor([p[1] if j==i else p.one() for j in range(N)]) for i in range(N)]
    lhs = framing_on_tensor((s[n].omega()((1-t)/(1-q)*sum(v[i]*X[i] for i in range(N))*epsilon([]))))
    rhs = s[n]((1-t)/(1-q)*sum(X[i]*(v[i]*u[i]+(1-q3)*sum(v[j]*u[j] for j in range(i+1,N))) for i in range(N))*epsilon([]))
    return lhs == rhs
```
