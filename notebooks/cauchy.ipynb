# Cauchy identity for Generalized Macdonald functions

This notebook verifies the **Cauchy identity** for Generalized Macdonald functions (GMPs).

## Statement

Let $N \geq 1$ and fix a total degree $d \geq 0$. Introduce independent sets of power-sum variables
$X_i = p_1^{(i)}$ for $i = 0, \ldots, N-1$, one per tensor factor. Define the *twisted* variables

$$\tilde{X}_i = X_i + (1 - q_3)\sum_{j=0}^{i-1} X_j,$$

which encode the spectral twist of the $N$-component scalar product. The Cauchy identity then reads:

$$\sum_{\boldsymbol{\lambda}\,:\,|\boldsymbol{\lambda}|=d}
\left(\prod_{k=0}^{N-1} b_{\lambda^{(k)}}\right)
\iota G_{\boldsymbol{\lambda}} \otimes G_{\boldsymbol{\lambda}}
= s_d\!\left[\frac{1-t}{1-q}\sum_{i=0}^{N-1} \tilde{X}_i \otimes X_i\right],$$

where:
- the sum runs over all $N$-tuples of partitions $\boldsymbol{\lambda} = (\lambda^{(0)},\ldots,\lambda^{(N-1)})$ of total weight $d$,
- $b(\lambda) = 1/\langle P_\lambda, P_\lambda \rangle_{q,t}$ is the inverse Macdonald norm (`blam`),
- $G_{\boldsymbol{\lambda}}$ is the GMP (`GMP`), $\iota G_{\boldsymbol{\lambda}}$ its *inverted* counterpart (`iGMP`), i.e. the GMP with partitions and weights in reversed order,
- $s_d$ is the complete homogeneous symmetric function $h_d$ written in the Schur basis,
- the right-hand side is a plethystic substitution into the degree-$d$ Schur function,
  evaluated on the bi-tensor argument $\sum_i \tilde{X}_i \otimes X_i$.

At $N=1$ this reduces to the classical Macdonald Cauchy identity
$\sum_{\lambda \vdash d} b_\lambda\, P_\lambda[x]\, P_\lambda[y] = s_d\bigl[\tfrac{1-t}{1-q}\, xy\bigr]$.

## Reference

Bourgine J.-E., Cassia L., Stoyan A. *Generalized Macdonald functions and quantum toroidal gl(1) algebra* (2025), [arXiv:2508.19704](https://arxiv.org/abs/2508.19704).

## Parameters

We verify the identity for $N = 3$ tensor factors at total degree $d = 2$.
The variables $X_i$ are the degree-1 generators of the $i$-th power-sum factor,
embedded in the full $N$-fold tensor product.

```python
N = 3  # number of tensor factors
d = 2  # total degree

# X[i] = p_1 in the i-th factor, 1 in all others
X = [tensor([p[1] if j == i else p.one() for j in range(N)]) for i in range(N)]
```

## Verification

**Left-hand side:** sum over all multi-partitions $\boldsymbol{\lambda}$ of total weight $d$,
weighted by $\prod_k b_{\lambda^{(k)}}$, of the bi-tensor $\bar{G}_{\boldsymbol{\lambda}} \otimes G_{\boldsymbol{\lambda}}$.

**Right-hand side:** the complete homogeneous function $s_d$ evaluated at the plethystic argument
$\tfrac{1-t}{1-q}\sum_{i} \tilde{X}_i \otimes X_i$,
where $\tilde{X}_i = X_i + (1-q_3)\sum_{j < i} X_j$ is the twisted variable for the $i$-th factor
(appearing in the left slot of the tensor product, corresponding to `iGMP`).

The cell evaluates to `True` if the identity holds.

```python
# Left-hand side: sum_lam b(lam) * iGMP(lam) x GMP(lam)
lhs = sum(
    prod(b_lambda(lam[i]) for i in range(N)) * tensor([iGMP(lam), GMP(lam)])
    for lam in mPartitions(N, d)
)

# Right-hand side: s_d( (1-t)/(1-q) * sum_i X_tilde[i] x X[i] )
# where X_tilde[i] = X[i] + (1-q3) * sum_{j<i} X[j]  (twisted variable, left slot)
rhs = s[d](
    (1 - t) / (1 - q) * sum(
        tensor([X[N - i - 1], (1 - q3) * sum(X[j] for j in range(i)) + X[i]])
        for i in range(N)
    )
)

lhs == rhs
```
