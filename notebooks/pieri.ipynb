# Pieri rules for Generalized Macdonald functions

This notebook verifies the **elementary Pieri rule** and its **dual** for
Generalized Macdonald functions (GMPs), as well as the single-factor vertex
coefficient identity that underpins them.

## Mathematical background

### Single-factor vertex coefficient

The function `psi_prime_PE(nu, lam)` is the vertex coefficient controlling
how the elementary symmetric function acts on a *single* Macdonald P-function.
It satisfies the classical N=1 Pieri rule:

$$\sum_{\nu} \psi'_{\nu/\lambda}\, P_\nu = e_m \cdot P_\lambda$$

where the sum runs over partitions $\nu$ of weight $|\lambda| + m$.
This is verified in cell 2 below for $1 \leq m, |\lambda| \leq 4$.

### N-component Pieri rule

For an $N$-tuple of partitions $\boldsymbol{\lambda}$, multiplication of
$P_{\boldsymbol{\lambda}}$ by $e_1[X_0 + \cdots + X_{N-1}]$ decomposes as:

$$e_1[X_0+\cdots+X_{N-1}]\cdot P_{\boldsymbol{\lambda}}
= \sum_{i=0}^{N-1}\sum_{\nu:\,\lambda^{(i)}\to\nu}
\alpha_N(i,\nu,\boldsymbol{\lambda})\,\psi'_{\nu/\lambda^{(i)}}\cdot
P_{\boldsymbol{\lambda}^{(i\to\nu)}}$$

where $\boldsymbol{\lambda}^{(i\to\nu)}$ denotes $\boldsymbol{\lambda}$ with
the $i$-th component replaced by $\nu$, and the inner sum runs over
partitions $\nu$ obtained from $\lambda^{(i)}$ by adding one box.
The factor $\alpha_N(i,\nu,\boldsymbol{\lambda})$ encodes the interaction
between component $i$ and the components after it.

### Dual Pieri rule

The dual rule controls the skewing of $P_{\boldsymbol{\nu}}$ by
$e_1\!\left[\frac{1-q}{1-t}\sum_i q_3^i X_i\right]$:

$$e_1\!\left[\frac{1-q}{1-t}\sum_i q_3^i \frac{\partial}{\partial X_i}\right]\cdot P_{\boldsymbol{\nu}}
= \sum_{\boldsymbol{\lambda}:\,|\boldsymbol{\lambda}|=|\boldsymbol{\nu}|-\square}
\prod_{i=0}^{N-1} \alpha_N^*(i,\boldsymbol{\nu},\lambda^{(i)})\,
\psi'^{\ast}_{\nu^{(i)}/\lambda^{(i)}}\cdot P_{\boldsymbol{\lambda}}$$

where the sum runs over multi-partitions reachable by removing one box
in total across all components.

## Single-factor check

We verify that `psi_prime_PE` correctly reproduces the classical N=1
elementary Macdonald Pieri rule.  This is a necessary condition for the
N-component rule to hold.

```python
# elementary Pieri rule:  sum_nu psi'(nu, lam) * P_nu = e_m * P_lam
print("Level one Pieri:",
    all(sum(psi_prime_PE(nu, lam) * McdP(nu) for nu in Partitions(sum(lam) + m)) == McdP([1]*m) * McdP(lam)
        for n in range(1, 5) for m in range(1, 5) for lam in Partitions(n)))
```

## N-component Pieri rule

`pieriTest(lam)` checks the elementary Pieri rule for the GMP $P_{\boldsymbol{\lambda}}$: it computes both sides and returns `True` if they are equal.

We verify it for all $N$-tuples of partitions of total weight $1 \leq d \leq 4$,
across $1 \leq N \leq 4$ components.

```python
print("Pieri rule:",
    all(pieriTest(lam)
        for N in range(1,5) for d in range(1, 5) for lam in mPartitions(N, d)))
```

## N-component dual Pieri rule

`pieriTestDual(nu)` checks the dual (skew) Pieri rule for $P_{\boldsymbol{\nu}}$.

We verify it for all $N$-tuples of partitions of total weight $1 \leq d \leq 4$,
across $1 \leq N \leq 4$ components.

```python
print("Dual Pieri rule:",
    all(pieriTestDual(nu)
        for N in range(1, 5) for d in range(1, 5) for nu in mPartitions(N, d)))
```
