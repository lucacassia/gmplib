# Conjecture 3.1

This notebook provides numerical evidence for **Conjecture 3.1** of

> J.-E. Bourgine, L. Cassia, A. Stoyan,
> *Generalized Macdonald functions and quantum toroidal gl(1) algebra* (2025),
> [arXiv:2508.19704](https://arxiv.org/abs/2508.19704).

## Statement

The conjecture is a higher-level generalisation of a classical identity,
which at level $N=1$ admits a proof via properties of (modified) Macdonald polynomials. At higher level $N > 1$
no proof is currently known; the cells below verify it computationally
for small values of $N$ and degree $n$.

For $N \geq 1$ and $n \geq 1$, introduce degree-1 generators
$X_i = p_1^{(i)}$ in the $N$-fold tensor product of power-sum algebras,
and let $\varepsilon_{\emptyset}=-\frac{1}{1-t}$ denote the principal specialisation
at the empty partition. The conjecture states:

$$\nabla(\boldsymbol{u})\cdot s_{[n]}\left[-\epsilon\frac{1-t}{1-q}\varepsilon_{\emptyset}\sum_{i=0}^{N-1} v_i\, X_i\right]
= s_{[n]}\!\left[\frac{1-t}{1-q}\varepsilon_{\emptyset}\sum_{i=0}^{N-1} X_i\,\bigl(v_i u_i + (1-q_3)\textstyle\sum_{j>i} v_j u_j\bigr)\right]$$

where $p_k[-\epsilon X] = -(-1)^kp_k[X]$ denotes the omega involution, $\nabla(\boldsymbol{u})$ is the framing operator
(`framing_on_tensor`), $u_i$ are the spectral parameters, and $v_i$ are
auxiliary parameters.

## Conjecture 3.1 — verification

`testConjecture(N, n)` returns `True` if the identity holds for the given
number of components $N$ and degree $n$. The trivial cases $N=0$ or $n=0$
return `True` immediately.

```python
def testConjecture(N, n):
    """Return True if Conjecture 3.1 holds for N components at degree n."""
    if n <= 0 or N <= 0:
        return True
    X = [tensor([p[1] if j == i else p.one() for j in range(N)]) for i in range(N)]
    lhs = framing_on_tensor(
        s[n].omega()((1-t)/(1-q) * sum(v[i]*X[i] for i in range(N)) * epsilon([]))
    )
    rhs = s[n](
        (1-t)/(1-q) * sum(
            X[i] * (v[i]*u[i] + (1-q3)*sum(v[j]*u[j] for j in range(i+1, N)))
            for i in range(N)
        ) * epsilon([])
    )
    return lhs == rhs
```

## Evaluation identity

An equivalent formulation is given by `testEvaluation` below, which
rewrites the conjecture as an evaluation identity for GMPs obtained
by expanding the plethystic exponentials via the Cauchy identity:

$$\mathrm{PE}\!\left[-z\sum_i u_i\,\chi_{\lambda^{(i)}}\right]\cdot
G_{\boldsymbol{\lambda}}\!\left[-\varepsilon_{\emptyset}\, q_3^i\right]
= G_{\boldsymbol{\lambda}}\!\left[-\varepsilon_{\emptyset}\,(q_3^i - z\, u_i)\right]$$

where the shorthand $G_{\boldsymbol{\lambda}}[\cdots]$ means evaluating
each tensor factor at the given argument. This corresponds to equation (4.87) in the paper.

```python
def testEvaluation(lam):
    """Return True if the GMP evaluation identity holds for the multi-partition lam."""
    N = len(lam)
    lhs = (PE(-z * sum(u[i]*chi2d(lam[i]) for i in range(N)))
           * evaluate_on_tensor(GMP(lam), [-epsilon([])*q3^i for i in range(N)]))
    rhs = evaluate_on_tensor(GMP(lam), [-epsilon([])*(q3^i - z*u[i]) for i in range(N)])
    return lhs == rhs
```

```python
print("Evaluation identity:",
    all(testEvaluation(lam) for N in range(1, 4) for d in range(4) for lam in mPartitions(N, d)))
```
