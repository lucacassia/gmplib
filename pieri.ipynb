```python
def pieriTestNew(lam):
    N = len(lam)
    X = [tensor([p[1] if j==i else p[[]] for j in range(N)]) for i in range(N)]
    lhs = e[1](sum(X))*GMP(lam)
    rhs = sum( prod(alpha_N(i,nu[i],lam)*psi_prime_PE(nu[i],lam[i]) for i in range(N)) * GMP(nu) for nu in pieri_set(1,lam))
    return lhs==rhs
```

```python
all(pieriTestNew(lam) for N in range(5) for n in range(1,5) for lam in mPartitions(N,n))
```
